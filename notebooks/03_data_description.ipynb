```python
import pandas as pd 
```

### Import Dataset 

```python
pv_df_musim_dingin = pd.read_excel("../datasets/raw/pv_musim_dingin.xlsx")
pv_df_musim_gugur = pd.read_excel("../datasets/raw/pv_musim_gugur.xlsx")
pv_df_musim_panas = pd.read_excel("../datasets/raw/pv_musim_panas.xlsx")
pv_df_musim_semi = pd.read_excel("../datasets/raw/pv_musim_semi.xlsx")
```

```python
wind_df_musim_dingin = pd.read_excel("../datasets/raw/wind_musim_dingin.xlsx")
wind_df_musim_gugur = pd.read_excel("../datasets/raw/wind_musim_gugur.xlsx")
wind_df_musim_panas = pd.read_excel("../datasets/raw/wind_musim_panas.xlsx")
wind_df_musim_semi = pd.read_excel("../datasets/raw/wind_musim_semi.xlsx")
```

### combine seasonal dataset 

```python
df_pv = pd.concat([
    pv_df_musim_dingin,
    pv_df_musim_semi,
    pv_df_musim_panas,
    pv_df_musim_gugur
], axis=0)
```

```python
df_pv.info()
```

```python
df_wind = pd.concat([
    wind_df_musim_dingin,
    wind_df_musim_semi,
    wind_df_musim_panas,
    wind_df_musim_gugur
], axis=0)
```

```python
df_wind.info()
```

### Rename name column

```python
kolom = [
    'timestamp',
    'time',
    'temperature',
    'humidity',
    'surface_radiation',
    'upper_atmospheric_radiation',  
    'output_pv', 
]

df_pv.columns = kolom
```

```python
df_pv.info()
```

```python
kolom = [
    'timestamp',
    'time',
    'air_density',
    'wind_velocity',
    'output_wind', 
]

df_wind.columns = kolom
```
