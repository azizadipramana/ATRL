```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import MinMaxScaler 
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.linear_model import Ridge
```

```python
df_train = pd.read_csv('../datasets/processed/train_lag.csv')
df_eval = pd.read_csv('../datasets/processed/eval_lag.csv')

cumulative_contribution = pd.read_csv('../result/cumulative_contribution.csv')
target = 'output_hybrid'
```

```python
base_feature = [
    'temperature', 
    'humidity', 
    'surface_radiation', 
    'upper_atmospheric_radiation',  
    'air_density', 
    'wind_velocity', 
    'hour', 
    'month'
] 

y_train = df_train[target]
y_test = df_eval[target]
```

### Evaluation

```python
sns.set_theme(style="whitegrid")
plt.rcParams.update({
    'font.size': 12,
    'axes.labelsize': 12,
    'axes.titlesize': 14,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'figure.titlesize': 16
})



def calculate_mape(y_true, y_pred):
    epsilon = 1e-10
    return np.mean(np.arctan(np.abs((y_true - y_pred) / (np.abs(y_true) + epsilon)))) * 100

list_model = {
    'RidgeRegression': Ridge(alpha=1.0),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'XGBoost': XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1),
    'LightGBM': LGBMRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1, verbose=-1), 
    'SVR': SVR(kernel='rbf', C=1.0, epsilon=0.1)
}

list_thresholds = [0.70, 0.75, 0.80, 0.85, 0.90, 0.95, 1.00]
all_records = []
for nama_model, regressor in list_model.items():
    print(f"{nama_model}")
    
    records = []
    
    for th in list_thresholds:
        if th == 1.00:
            lag_lolos = cumulative_contribution['Feature'].tolist()
        else:
            idx_cutoff = (cumulative_contribution['Cumulative_Contribution'] >= th).idxmax()
            lag_lolos = cumulative_contribution['Feature'].iloc[:idx_cutoff + 1].tolist()
        
        fitur_aktif_total = base_feature + lag_lolos
        num_lag_only = len(lag_lolos) 
        
        X_train = df_train[fitur_aktif_total]
        X_test = df_eval[fitur_aktif_total]
        
        scaler_X = MinMaxScaler(feature_range=(0, 1))
        scaler_y = MinMaxScaler(feature_range=(0, 1))
        
        X_train = scaler_X.fit_transform(X_train)
        X_test = scaler_X.transform(X_test)
        
        y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
        
        regressor.fit(X_train, y_train_scaled)

        preds_scaled = regressor.predict(X_test)
        preds = scaler_y.inverse_transform(preds_scaled.reshape(-1, 1)).flatten()
        
        rmse = np.sqrt(mean_squared_error(y_test, preds))
        mae = mean_absolute_error(y_test, preds)
        mape = calculate_mape(y_test, preds)
        r2 = r2_score(y_test, preds)

        records.append({
            'Model': nama_model, # Tambahkan key Model agar mudah di-grouping nanti
            'Threshold': th * 100,
            'Num_Lag_Features': num_lag_only, 
            'RMSE': rmse,
            'MAE': mae, 
            'MAPE': mape, 
            'R2': r2
        })
        all_records.append(records[-1])
    df_plot = pd.DataFrame(records)

    print(df_plot.to_string(index=False, formatters={
        'RMSE': '{:,.4f}'.format, 
        'MAE': '{:,.4f}'.format,
        'MAPE': '{:,.2f}'.format,
        'R2': '{:,.4f}'.format
    }))

    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    sns.lineplot(
        data=df_plot, x='Threshold', y='RMSE', 
        marker='o', markersize=8, color='#d9534f', linewidth=2.5, ax=axes[0]
    )
    min_rmse_idx = df_plot['RMSE'].idxmin()
    opt_th = df_plot.loc[min_rmse_idx, 'Threshold']
    opt_rmse = df_plot.loc[min_rmse_idx, 'RMSE']

    axes[0].plot(opt_th, opt_rmse, marker='X', color='black', markersize=12, label=f'Sweet Spot ({opt_th:.0f}%)')
    axes[0].set_title(f"Grafik 1: Threshold vs RMSE ({nama_model})")
    axes[0].set_xlabel("Contribution Threshold (%)")
    axes[0].set_ylabel("Forecasting Error (RMSE)")
    axes[0].set_xticks(df_plot['Threshold'])
    axes[0].legend(loc='upper right')
    axes[0].grid(True, linestyle='--', alpha=0.7)

    sns.lineplot(
        data=df_plot, x='Threshold', y='Num_Lag_Features', 
        marker='s', markersize=8, color='#0275d8', linewidth=2.5, ax=axes[1]
    )
    opt_lag_features = df_plot.loc[min_rmse_idx, 'Num_Lag_Features']
    axes[1].plot(opt_th, opt_lag_features, marker='X', color='black', markersize=12, label=f'Selected: {opt_lag_features} Lags')

    axes[1].set_title(f"Grafik 2: Threshold vs Complexity ({nama_model})")
    axes[1].set_xlabel("Contribution Threshold (%)")
    axes[1].set_ylabel("Number of Selected Lag Features (#)")
    axes[1].set_xticks(df_plot['Threshold'])
    axes[1].legend(loc='upper left')
    axes[1].grid(True, linestyle='--', alpha=0.7)

    plt.tight_layout()
    
    save_file_name = f'../result/figure/compactness_tradeoff_{nama_model}.png'
    plt.savefig(save_file_name, dpi=300, bbox_inches='tight')
    print(f"\n[Success] The visualization is saved in: '{save_file_name}'")
    
    plt.show()
```

```python
df_all_results = pd.DataFrame(all_records)

best_summary_records = []

for nama_model in list_model.keys():
    df_model_only = df_all_results[df_all_results['Model'] == nama_model]
    
    if not df_model_only.empty:
        idx_terbaik = df_model_only['RMSE'].idxmin()
        row_terbaik = df_model_only.loc[idx_terbaik].to_dict()
        
        best_summary_records.append(row_terbaik)

df_final_summary = pd.DataFrame(best_summary_records)

print("Summary of the Best Lag Configuration (Sweet Spot)\n")
print(df_final_summary.to_string(index=False, columns=[
    'Model', 'Threshold', 'Num_Lag_Features', 'RMSE', 'MAE', 'MAPE', 'R2'
], formatters={
    'Threshold': '{:.0f}%'.format,
    'RMSE': '{:,.4f}'.format, 
    'MAE': '{:,.4f}'.format,
    'MAPE': '{:,.2f}'.format, 
    'R2': '{:,.4f}'.format
}))
```

```python
sns.set_theme(style="whitegrid")
plt.rcParams.update({'font.size': 12, 'axes.titlesize': 14})

y_train = df_train[target]
y_test = df_eval[target]

def calculate_mape(y_true, y_pred):
    epsilon = 1e-10
    return np.mean(np.abs((y_true - y_pred) / (y_true + epsilon))) * 100

list_model = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    'XGBoost': XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1),
    'LightGBM': LGBMRegressor(n_estimators=150, learning_rate=0.05, max_depth=6, random_state=42, n_jobs=-1, verbose=-1), 
    'SVR': SVR(kernel='rbf', C=1.0, epsilon=0.1)
}

list_thresholds = [0.70, 0.75, 0.80, 0.85, 0.90, 0.95, 1.00]
all_records = [] 

for nama_model, regressor in list_model.items():
    print(f"Training: {nama_model}...")
    
    for th in list_thresholds:
        if th == 1.00:
            lag_pass = cumulative_contribution['Feature'].tolist()
        else:
            idx_cutoff = (cumulative_contribution['Cumulative_Contribution'] >= th).idxmax()
            lag_pass = cumulative_contribution['Feature'].iloc[:idx_cutoff + 1].tolist()
        
        feature_used = base_feature + lag_pass
        
        X_train = df_train[feature_used]
        X_test = df_eval[feature_used]
        
        scaler_X = MinMaxScaler()
        scaler_y = MinMaxScaler()
        
        X_train_scaled = scaler_X.fit_transform(X_train)
        X_test_scaled = scaler_X.transform(X_test)
        y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1)).flatten()
        
        regressor.fit(X_train_scaled, y_train_scaled)
        preds = scaler_y.inverse_transform(regressor.predict(X_test_scaled).reshape(-1, 1)).flatten()
        
        all_records.append({
            'Model': nama_model,
            'Threshold': th * 100,
            'Num_Lag_Features': len(lag_pass),
            'RMSE': np.sqrt(mean_squared_error(y_test, preds))
        })

# Visualization
df_all = pd.DataFrame(all_records)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Figure 1: RMSE
sns.lineplot(data=df_all, x='Threshold', y='RMSE', hue='Model', marker='o', linewidth=2, ax=axes[0])
axes[0].set_title("Perbandingan RMSE Semua Model")
axes[0].grid(True, linestyle='--', alpha=0.7)

# Figure 2: Kompleksitas
sns.lineplot(data=df_all, x='Threshold', y='Num_Lag_Features', hue='Model', marker='s', linewidth=2, ax=axes[1])
axes[1].set_title("Perbandingan Jumlah Fitur (Kompleksitas)")
axes[1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('../result/figure/perbandingan_semua_model.png', dpi=150)
plt.show()
```
