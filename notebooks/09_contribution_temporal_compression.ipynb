```python
import sys
sys.path.append("..")
import pandas as pd 
from src.feature_selection import (
    calculate_cumulative_contribution,
    select_features_by_contribution
)
```

### Cumulative Contribution

```python
ensemble_df = pd.read_csv("../result/ensemble_selected.csv")

ensemble_df = calculate_cumulative_contribution(ensemble_df, score_column="Ensemble_Score")
selected_df, selected_features = (
    select_features_by_contribution(
        ensemble_df,
        threshold=0.90
    )
)

ensemble_df.to_csv("../result/cumulative_contribution.csv", index=False)
print(
    ensemble_df[["Feature", "Ensemble_Score", "Cumulative_Contribution"]]
    .head(50)
    .to_string(
        index=False,
        formatters={
            "Ensemble_Score": "{:,.4f}".format,
            "Cumulative_Contribution": "{:.2%}".format
        }
    )
)

print("\nSelected Features:")
print(selected_features)

print(
    f"\nSelected Features Count: "
    f"{len(selected_features)}"
)

print(
    f"Original Features Count: "
    f"{len(ensemble_df)}"
)
```
