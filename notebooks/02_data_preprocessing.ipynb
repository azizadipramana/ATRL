```python
# 1. Ubah kolom timestamp masing-masing dataset menjadi objek 'set'
# Pastikan nama kolomnya sesuai (misal: 'timestamp')
set_pv = set(pv_df['timestamp'])
set_wind = set(wind_df['timestamp'])

# 2. Lakukan pengurangan set (set subtraction) untuk mencari selisihnya
missing_in_wind = set_pv - set_wind  # Ada di PV, tapi ga ada di Wind
missing_in_pv = set_wind - set_pv    # Ada di Wind, tapi ga ada di PV

print(f"Jumlah timestamp di PV tapi ga ada di Wind : {len(missing_in_wind)}")
print(f"Jumlah timestamp di Wind tapi ga ada di PV : {len(missing_in_pv)}")

print("-" * 50)

# 3. Pengecekan menggunakan 'if' yang aman (karena objeknya berupa set / list)
if missing_in_pv:
    print("Contoh data yang di PV ga ada, tapi di Wind ada (5 teratas):")
    # Urutkan biar enak dibaca
    for ts in sorted(list(missing_in_pv))[:5]:
        print(ts)
else:
    print("Aman! Semua timestamp di data Wind ada di data PV.")

print("-" * 50)

if missing_in_wind:
    print("Contoh data yang di Wind ga ada, tapi di PV ada (5 teratas):")
    for ts in sorted(list(missing_in_wind))[:5]:
        print(ts)
else:
    print("Aman! Semua timestamp di data PV ada di data Wind.")
```
